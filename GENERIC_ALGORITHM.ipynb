```python
def get_max(cromolenght):
    return((2**cromolenght)-1)


def initial_pop(population, cromolenght):
    import random
    random_val = []
    for i in range(population):
        random_val.append(str(bin(random.randint(
            0, get_max(cromolenght)))[2:]).zfill(cromolenght))
    return(random_val)


def fitness_function(random_val):
    fitness_result = []
    for i in range(0, len(random_val)):
        fitness_result.append(int(random_val[i], 2))
    return(fitness_result)


def selection(random_val, fitness_result):
    min_index = fitness_result.index(min(fitness_result))
    max_index = fitness_result.index(max(fitness_result))
    random_val[min_index] = random_val[max_index]
    return(random_val)


def cross_over(random_val):
    for i in range(0, len(random_val)-1, 2):
        temp_1 = (random_val[i])[:4]+(random_val[i+1])[4:]
        temp_2 = (random_val[i+1])[:4]+(random_val[i])[4:]
        random_val[i] = temp_1
        random_val[i+1] = temp_2
    return(random_val)


def mutation(random_val):
    for i in range(0, len(random_val)):
        random_val[i] = random_val[i].replace('0', '1', 1)
    return(random_val)


def Max_condition_not_true(random_val, cromolenght):
    fitness_result = fitness_function(random_val)
    if(get_max(cromolenght) in fitness_result):
        return(False)

    return(True)


def eval():
    population = 10
    cromolenght = 255
    random_val = initial_pop(population, cromolenght)
    print("Initial random numbers are ", random_val, "\n\n\n\n")
    fitness_result = []
    gen = -1
    while(Max_condition_not_true(random_val, cromolenght)):
        fitness_result = fitness_function(random_val)
        random_val = selection(random_val, fitness_result)
        random_val = cross_over(random_val)
        random_val = mutation(random_val)
        print(" mutation are : ", random_val)
        gen = gen+1
        print("\n\n\n\n")
    print("gen are : ", gen)


eval()
```
